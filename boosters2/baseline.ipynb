## Train

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
def months_delation(raw):
    delation = raw.field3 - raw.field2
    if delation < 0:
        return 12 + delation
    return delation
```

```python
def new_year(raw):
    delation = raw.field3 - raw.field2
    if delation < 0:
        return 1
    return 0
```

```python
train['months_delation'] = train.apply(months_delation, axis=1)
train['new_year'] = train.apply(new_year, axis=1)
```

```python
train.months_delation.value_counts()
```

```python
train.new_year.value_counts()
```

```python
X_train, X_val, y_train, y_val = train_test_split(train.drop('orderid', axis=1), target, random_state=5)
```

```python
categorical_columns = [
    'userid', 'field2', 'field3', 'field29', 'field18', 'field20', 'field27', 'field11', 'field23'
]
```

```python
parameters = {
#     'learning_rate': [1e-4, 1e-3, 1e-2],
    'depth': [1, 5, 10, None],
#     'min_data_in_leaf': [1, 5, 10, 20],
#     'num_leaves': [8, 10, 15, 20, None],
    'iterations': [10, 20, 40, 70, 100]
}
```

```python
model = CatBoostClassifier(
    loss_function='Logloss',
    eval_metric='AUC',
    verbose=2,
    cat_features=categorical_columns
)
```

```python
clf = GridSearchCV(model, parameters, verbose=2)
```

```python
clf.fit(train.drop('orderid', axis=1), target)
```

```python
clf.best_params_
```

```python
clf.best_estimator_
```

```python
test['months_delation'] = test.apply(months_delation, axis=1)
test['new_year'] = test.apply(new_year, axis=1)
```

```python
prediction = clf.best_estimator_.predict_proba(test.drop('orderid', axis=1))
```

```python
test['proba'] = prediction[:, 1]
```

```python
baseline = test[['orderid', 'proba']]
```

```python
baseline.to_csv('baseline_gridsearch_depth_iterations.csv', index=False)
```

```python
model = CatBoostClassifier(
    iterations=200,
    loss_function='Logloss',
    eval_metric='AUC'
)
```

```python
model.fit(X_train, y_train,
          cat_features=categorical_columns,
          eval_set=(X_val, y_val),
          verbose=2)
```

```python
test['months_delation'] = test.apply(months_delation, axis=1)
test['new_year'] = test.apply(new_year, axis=1)
```

```python
prediction = model.predict_proba(test.drop('orderid', axis=1))
```

```python
test['proba'] = prediction[:, 1]
```

```python
baseline = test[['orderid', 'proba']]
```
